# Visualization with Matplotlib

Matplotlib is a plotting library for Python. It is used along with NumPy to provide an environment that is an effective open source alternative for MatLab. It can also be used with graphics toolkits like PyQt and wxPython.

Matplotlib module was first written by John D. Hunter. Since 2012, Michael Droettboom is the principal developer. Currently, Matplotlib ver. 1.5.1 is the stable version available. The package is available in binary distribution as well as in the source code form on www.matplotlib.org.

Conventionally, the package is imported into the Python script by adding the following statement âˆ’

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

f = open(r'C:\Users\Rutvik Mehta\Desktop\Python workshop Exam\mbox-short.txt', 'r')
freq_clients = {}
emails = list()
data = f.readlines()

for line in data:
    if line.startswith("From:"):
        atpos = line.find("@")
        spos = line.find(" ",atpos)
        email = line[atpos+1:spos] 
        emails.append(email)
for domain in emails:
    if domain not in freq_clients:
        freq_clients[domain] = 1
    else:
        freq_clients[domain] = freq_clients[domain] + 1

#print freq_clients
plt.plot(range(len(freq_clients.items())),list(freq_clients.values()))
plt.title("email count")
plt.xticks(range(len(freq_clients.items())),list(freq_clients.keys()))
plt.xlabel("domains")
plt.ylabel("count")
plt.grid(True, color = "k")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
f = open(r'C:\Users\Rutvik Mehta\Desktop\Python workshop Exam\mbox-short.txt', 'r')
freq_clients = {}
emails = list()
data = f.readlines()

for line in data:
    if line.startswith("From:"):
        atpos = line.find("@")
        spos = line.find(" ",atpos)
        email = line[atpos+1:spos] 
        emails.append(email)
for domain in emails:
    if domain not in freq_clients:
        freq_clients[domain] = 1
    else:
        freq_clients[domain] = freq_clients[domain] + 1

#print freq_clients


plt.bar(range(len(freq_clients.items())),list(freq_clients.values()),color=['black', 'red', 'green', 'blue', 'cyan'])
plt.title("email count")
plt.xticks(range(len(freq_clients.items())),list(freq_clients.keys()))
plt.xlabel("domains")
plt.ylabel("count")
plt.show()
```
